```python
import math
import random
import matplotlib.pyplot as plt
```

```python
# hexagon length 2*a with centre at (0,0). The 2 cancels out to make the math easier
a = 5.0
hexagon = [[-a, -math.sqrt(3) * a], [a, -math.sqrt(3) * a], [2*a, 0], [a, math.sqrt(3) * a], [-a, math.sqrt(3) * a], [-2*a, 0]]
```

```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
#create hexagon
polygon = Polygon(hexagon) # create hexagon
```
